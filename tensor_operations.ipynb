# Math Operations
- Add, substract, multiply, divide, reaminders, exponents
- shorthand and longhand
- reassignment

```python
import numpy as np
```

```python
ta = torch.tensor([1,2,3,4])
tb = torch.tensor([5,6,7,8])
print(ta)
print(tb)
```

```python
# Addition
ta+tb
```

```python
# Addition longhand
torch.add(ta,tb)
```

```python
# Subtraction
ta - tb
```

```python
# subtraction Longhand
# sub() is same with subtract()
torch.sub(ta, tb)
torch.subtract(ta, tb)
```

```python
# Multiplication
ta * tb
```

```python
torch.multiply(ta, tb)
```

```python
# Division
ta / tb
```

```python
torch.divide(ta,tb)
```

```python
# Remainder
ta % tb
```

```python
torch.remainder(ta,tb)
```

```python
# Power
ta ** tb
```

```python
torch.pow(ta,tb)
```

```python
# annother way to add
ta.add(tb)
```

```python
# Reassignment
# ta = ta + tb
ta.add_(tb)
```
